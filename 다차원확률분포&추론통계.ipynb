```python
import numpy as np
import scipy as sp
import pandas as pd
from pandas import Series, DataFrame

import matplotlib.pyplot as plt
import matplotlib as mpl
import seaborn as sns
import datetime
%precision 3
import scipy.stats as st
from scipy.stats import multivariate_normal
from mpl_toolkits.mplot3d import Axes3D

np.random.seed(0)
```

```python
x, y = np.mgrid[10:100:2, 10:100:2]

pos = np.empty(x.shape + (2,))

pos[:, :, 0] = x
pos[:, :, 1] = y
rv = multivariate_normal([50,50],[[100,0], [0,100]])

z = rv.pdf(pos)
```

```python
fig = plt.figure(dpi=100)

ax = Axes3D(fig)
ax.plot_wireframe(x, y, z)

ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('f(x, y)')

ax.ticklabel_format(style='sci', axis='z', scilimits=(0,1))
```

```python
calc_times = 1000
sample_array = np.array([1,2,3,4,5,6])
number_cnt = np.arange(1, calc_times + 1)

for i in range(4):
    p = np.random.choice(sample_array, calc_times).cumsum()
    plt.plot( p / number_cnt)
    plt.grid(True)
```

```python
def function_central_theory(N):
    sample_array = np.array([1,2,3,4,5,6])
    number_cnt = np.arange(1, N + 1)
    print(number_cnt)
    mean_array = np.array([])
    
    for i in range(1000):
        cum_variables = np.random.choice(sample_array, N).cumsum() * 1.0
        #print(cum_variables)
        mean_array = np.append(mean_array, cum_variables[N - 1 ] / N)
        
    plt.hist(mean_array)
    plt.grid(True)
```

```python
function_central_theory(30)
```

```python
function_central_theory(10)
```

```python
function_central_theory(20)
```
