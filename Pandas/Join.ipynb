```python
import numpy as np
import numpy.random as random
import scipy as sp
import pandas as pd
from pandas import Series, DataFrame

import matplotlib.pyplot as plt
import matplotlib as mpl
import seaborn as sns
%matplotlib inline

%precision 3
```

```python
data1 = {
    'id' : ['100', '101', '102', '103', '104', '106', '108', '110', '111', '113'],
    'city' : ['Seoul', 'Busan', 'Daegu', 'Gangreung', 'Seoul', 'Seoul', 'Busan', 'Daegu', 'Gangreung', 'Seoul'],
    'birth_year' : [1990, 1989, 1992, 1997, 1982, 1991, 1988, 1990, 1995, 1981],
    'name' : ['Junho', 'Heejin', 'Mijung', 'MInho', 'Steeve', 'Mina', 'Sumi', 'Minsu', 'Jinhee', 'Daeho']
}

df1 = DataFrame(data1)
df1
```

```python
data2 = {
    'id' : ['100', '101', '102', '105', '107'],
    'math' : [50, 43, 33, 76, 98],
    'english' : [90, 30, 20, 50, 30],
    'sex' : ['M', 'F', 'F', 'M', 'M'],
    'index_num' : [0,1,2,3,4]
}
df2 = DataFrame(data2)
df2
```

```python
print('INNER JOIN Table')
pd.merge(df1, df2, on = 'id')
```

```python
print('OUTER JOIN Table')
pd.merge(df1, df2, how='outer')
```

```python
pd.merge(df1, df2, left_index= True, right_on= 'index_num')
```

```python
pd.merge(df1, df2, how='left')
```

```python
pd.merge(df1, df2, how='right')
```
