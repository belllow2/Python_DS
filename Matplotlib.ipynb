```python
np.random.seed(0)

x = np.random.randn(30)
y = np.sin(x) + np.random.randn(30)

plt.figure(figsize=(10,10))

plt.plot(x,y,'o')
#plt.scatter(x,y)

plt.title('Name')
plt.xlabel('X')
plt.ylabel('Y')
plt.grid(True)
```

```python
np.random.seed(0)

numpy_data_x = np.arange(1000)
numpy_random_data_y = np. random.randn(1000).cumsum()

plt.figure(figsize=(10,10))
plt.plot(numpy_data_x,numpy_random_data_y, label='Y')
plt.legend()
plt.xlabel('X')
plt.ylabel('Y')
plt.grid(True)
```

```python
# 

data = np.random.uniform(0.0, 1.0, 10000)
print(type(data))
data2 = np.random.uniform(0.0, 1.0, 10000)
theta = np.linspace(0, 2*np.pi, 100)
radius = 1.0
x = radius*np.cos(theta)
y = radius*np.sin(theta)
plt.figure(figsize=(12,6))
outside_x = []
outside_y = []
inside_x = []
inside_y = []

n = 0
for i in range(len(data)):
    if math.hypot(data[i],data2[i]) < 1:
        inside_x.append(data[i])
        inside_y.append(data2[i])
        n += 1
    else:
        outside_x.append(data[i])
        outside_y.append(data2[i])
        
plt.plot(x,y)        
plt.plot(inside_x,inside_y, color="r")
plt.plot(outside_x,outside_y, color="b")
print(n)
print(" : ", (4.0 * n) / len(data))
```
